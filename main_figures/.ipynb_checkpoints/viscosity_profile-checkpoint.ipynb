```python
def cartesian_to_spherical(x, y, z):
    """
    Takes an x, y, z and converts it to spherical coordinates. Returns r, theta, phi
    """
    r = np.sqrt(x**2 + y**2 + z**2)
    theta = 90 - np.rad2deg( np.arccos( z / (np.sqrt(x**2 + y**2 + z**2)) ) )
    phi =  np.sign(y) * np.rad2deg(np.arccos( x / np.sqrt(x**2 + y**2) ))
    phi[np.where(phi < 0)] = phi[np.where(phi < 0)] + 360
    phi[np.where(phi == 0)] = 180
    
    return r, phi, theta

def spherical_to_global_cartesian(r, phi, theta):
    """
    Takes spherical coordinates r, theta, and phi and converts to Cartesian coordinates.
    Returns x, y, z
    """
    x = r * np.sin(np.deg2rad(90 - theta)) * np.cos(np.deg2rad(phi))
    y = r * np.sin(np.deg2rad(90 - theta)) * np.sin(np.deg2rad(phi))
    z = r * np.cos(np.deg2rad(90 - theta))
    
    return x, y, z
```

```python
def plate_cooling(T0, T1, z, zlo, t, kappa, n, zmax):
    dT = T1 - T0
    summation = 0
    exp_fac = -kappa * np.pi**2 * t / zlo**2
    sin_fac = np.pi / zlo
    if z > zlo:
        T = T1
    else:
        for m in range(1, n + 1):
            summation = summation + 1/m*np.exp(m**2 * exp_fac) * np.sin(m * (z) * sin_fac)
        T = T0 + dT * ( (z) / zlo + 2/np.pi * summation)
    return T

PS_T0 = 273
PS_T1 = 1623

z = np.linspace(0, 150e3, 10000)
PS_zlo = 125e3
s2yr = 60 * 60 * 24 * 365
t_vals = np.array([130e6]) * s2yr
labels = np.array(['90', '60', '40', '20'])
kappa = 0.8e-6
n = 5
zmax = np.max(z)
Temp = np.zeros(len(z))
Pressure = np.zeros(len(z))

for k in range(len(z)):
    Temp[k] = plate_cooling(PS_T0, PS_T1, z[k], PS_zlo, t_vals[0], kappa, n, zmax)
    Pressure[k] = 9.81 * 3300 * z[k] + 5e9
```

```python
n_diff = 1

A_diff_d = 4.5e-15
A_diff_w = 3e-15

E_diff_d = 375e3
E_diff_w = 335e3

V_diff_d = 8.2e-6
V_diff_w = 4e-6
grain_size = 1e-3
m_diff = 3

n_disl = 3.5

A_disl_d = 7.4e-15
A_disl_w = 2.4e-14

E_disl_d = 530e3
E_disl_w = 480e3

V_disl_d = 14e-16
V_disl_w = 11e-16
m_disl = 0

R = 8.3141
d = 1e-3
epsilon_ii = 1e-16

min_visc = 1e18
max_visc = 1e24

viscosity_diff_d = 1/2 * A_diff_d**(-1) * epsilon_ii**(0) * d**(3) * \
                   np.exp((E_diff_d + Pressure * V_diff_d) / R / Temp)

viscosity_disl_d = 1/2 * A_disl_d**(-1/n_disl) * epsilon_ii**((1-n_disl)/n_disl) * d**(0) * \
                   np.exp((E_disl_d + Pressure * V_disl_d) / R / Temp / n_disl)

dry_viscosity = (viscosity_diff_d * viscosity_disl_d) / (viscosity_diff_d + viscosity_disl_d)

dry_viscosity[np.where(dry_viscosity <= min_visc)] = min_visc
dry_viscosity[np.where(dry_viscosity >= max_visc)] = max_visc

viscosity_diff_w = 1/2 * A_diff_w**(-1) * epsilon_ii**(0) * d**(3) * \
                   np.exp((E_diff_w + Pressure * V_diff_w) / R / Temp)

viscosity_disl_w = 1/2 * A_disl_w**(-1/n_disl) * epsilon_ii**((1-n_disl)/n_disl) * d**(0) * \
                   np.exp((E_disl_w + Pressure * V_disl_w) / R / Temp / n_disl)

wet_viscosity = (viscosity_diff_w * viscosity_disl_w) / (viscosity_diff_w + viscosity_disl_w)

wet_viscosity[np.where(wet_viscosity <= min_visc)] = min_visc
wet_viscosity[np.where(wet_viscosity >= max_visc)] = max_visc

# wet_crust = 
```

```python
n_diff = 1

A_diff_d = 4.5e-15
A_diff_w = 3e-15

E_diff_d = 375e3
E_diff_w = 335e3

V_diff_d = 8.2e-6
V_diff_w = 4e-6
grain_size = 1e-3
m_diff = 3

n_disl = 3.5

A_disl_d = 7.4e-15
A_disl_w = 2.4e-14

E_disl_d = 530e3
E_disl_w = 480e3

V_disl_d = 14e-16
V_disl_w = 11e-16
m_disl = 0

R = 8.3141
d = 1e-3
epsilon_ii = 1e-17

min_visc = 1e19
max_visc = 1e24

TEMPERATURE = TEMP[profile_indices]

viscosity_diff_d = 1/2 * A_diff_d**(-1) * epsilon_ii**(0) * d**(3) * \
                   np.exp((E_diff_d + PRESSURE * V_diff_d) / R / TEMPERATURE)

viscosity_disl_d = 1/2 * A_disl_d**(-1/n_disl) * epsilon_ii**((1-n_disl)/n_disl) * d**(0) * \
                   np.exp((E_disl_d + PRESSURE * V_disl_d) / R / TEMPERATURE / n_disl)

dry_viscosity = (viscosity_diff_d * viscosity_disl_d) / (viscosity_diff_d + viscosity_disl_d)

dry_viscosity[np.where(dry_viscosity <= min_visc)] = min_visc
dry_viscosity[np.where(dry_viscosity >= max_visc)] = max_visc

viscosity_diff_w = 1/2 * A_diff_w**(-1) * epsilon_ii**(0) * d**(3) * \
                   np.exp((E_diff_w + PRESSURE * V_diff_w) / R / TEMPERATURE)

viscosity_disl_w = 1/2 * A_disl_w**(-1/n_disl) * epsilon_ii**((1-n_disl)/n_disl) * d**(0) * \
                   np.exp((E_disl_w + PRESSURE * V_disl_w) / R / TEMPERATURE / n_disl)

wet_viscosity = (viscosity_diff_w * viscosity_disl_w) / (viscosity_diff_w + viscosity_disl_w)

wet_viscosity[np.where(wet_viscosity <= min_visc)] = min_visc
wet_viscosity[np.where(wet_viscosity >= max_visc)] = max_visc
```
