## Calculate arc length of slab under North Islands

```python
north_island_prof = np.loadtxt(fname='SZ_data/kermadec/fix_slab_tear/260_contour/15prof_10spacing/cartesian/profile_00165.xyz')
north_island_prof_x = north_island_prof[:, 0] - north_island_prof[:, 0][0]
north_island_prof_y = north_island_prof[:, 1] - north_island_prof[:, 1][0]
north_island_prof_z = north_island_prof[:, 2]
```

```python
import numpy as np
beta = 0.3
eta_eff = 1e20
eta_ve = 5e19
print( (eta_eff*eta_ve/(eta_eff + eta_ve))**(1 - beta) * np.min(np.array([eta_eff, eta_ve]))**beta)
```

```python
arc_length = 0
for i in range(len(north_island_prof_x) - 1):
    xy_proj = np.sqrt(north_island_prof_y**2 + north_island_prof_x**2)[i + 1] - np.sqrt(north_island_prof_y**2 + north_island_prof_x**2)[i]
    z_change = north_island_prof_z[i + 1] - north_island_prof_z[i]
    arc_length += np.sqrt(xy_proj**2 + z_change**2)
print(arc_length)
```
